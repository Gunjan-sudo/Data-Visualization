```python
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns 
import pandas as pd
```

```python
fifa=pd.read_csv('players_20.csv')
```

```python
fifa.head()
```

```python
plt.figure(figsize=(6,6))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color='g')
plt.show()
```

```python
player_salary=fifa[['short_name','weight_kg']]
```

```python
player_salary.head()
```

```python
player_salary=player_salary.sort_values(by=['weight_kg'],ascending=False)
player_salary.head()
```

```python
plt.figure(figsize=(6,6))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['weight_kg'])[0:5],color=['green','red','yellow','purple','blue'])
plt.show()
```
