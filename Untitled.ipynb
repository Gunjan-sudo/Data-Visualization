```python
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
ip2=pd.read_csv("ipl23.csv")
```

```python
ip2.head()
```

```python
batting_second=ip2[ip2["win_by_wickets"]!=0]
batting_second.head()
```

```python
#making a histogram of the win by wickets i.e the batting_second team winner 
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
#The most matches won by the teams when batting at second innings
batting_second['winner'].value_counts()
```

```python
#bar graph indicating the top three teams won the match after batting at the second position 
plt.figure(figsize=(7,7))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=["yellow","blue","purple"])
plt.show()
```

```python
#The pie chart indicating the teams who have the won the no of percentage matches after batting at the second position 
plt.figure(figsize=(9,9))
plt.pie(list(batting_second['winner'].value_counts()), labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1ff%%')
plt.show()
```
